```python
sales['Unit_Cost'].max()
```

```python
sales['Unit_Cost'].plot(kind='box', vert=False, figsize=(14,6))   #chart kind box with value 'Unit_Cost'
```

```python
ax = sales['Unit_Cost'].plot(kind='kde', figsize=(14,6))  #kind kde chart describes the mean and median in the sales.
ax.axvline(sales['Unit_Cost'].mean(), color='red')
ax.axvline(sales['Unit_Cost'].median(), color='green')
```

```python
ax = sales['Unit_Cost'].plot(kind='hist', figsize=(14,6))  #kind histogram chart describes  the number sales and quantity.
ax.set_ylabel('Number Sales')
ax.set_xlabel('Dollars')
```
